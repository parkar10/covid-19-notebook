```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
DATA_PATH = '../data/covid19_india/'
```

```python
testing_facilities = pd.read_csv(f"{DATA_PATH}ICMRTestingDetails.csv")
```

```python
testing_facilities.head(5)
```

```python
del testing_facilities["SNo"]
```

```python
testing_facilities.fillna("0", inplace=True)
testing_facilities.columns
```

```python
plt.title("Testing increased")
plt.ylabel("Testing")
plt.xlabel("Date")
plt.style.use('ggplot')
plt.plot(testing_facilities["DateTime"],testing_facilities["TotalSamplesTested"],color="y")
plt.plot(testing_facilities["DateTime"],testing_facilities["TotalIndividualsTested"],color="b")
```

```python
plt.title("Cases")
plt.ylabel("Cases")
plt.xlabel("Date")
plt.style.use('ggplot')
plt.plot(testing_facilities["DateTime"],testing_facilities["TotalPositiveCases"],color="y")
```

```python
plt.title("Testing vs Cases")
plt.ylabel("Cases")
plt.xlabel("Testing")
plt.style.use('ggplot')
plt.plot(testing_facilities["TotalSamplesTested"],testing_facilities["TotalPositiveCases"],color="y")
```
