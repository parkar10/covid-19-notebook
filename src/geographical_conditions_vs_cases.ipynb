```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
import os
```

```python
DATA_PATH = '../data/Midas/'
os.listdir(DATA_PATH)
```

```python
geo_df = pd.read_csv(f"{DATA_PATH}data_from_papers/dataset_luo-et-al_202002/R0_absolute_humidity_Luo_et_al.2020.csv")
```

```python
geo_df.head(5)
```

```python
geo_df.fillna("0", inplace=True)
geo_df.columns
```

```python
geo_df_sorted = geo_df.astype({"Absolute_Humidity":"double","R0_mean":"double"}).sort_values(by=["Temperature_Mean"])
plt.title("R0 relation")
plt.ylabel("R0")
plt.xlabel("Temperature_Mean")
plt.style.use('ggplot')
plt.scatter(geo_df_sorted["Temperature_Mean"], geo_df_sorted["R0_mean"], color="y")
plt.scatter(geo_df_sorted["Temperature_Mean"], geo_df_sorted["R0_confidence_interval_Upper"], color="r")
plt.scatter(geo_df_sorted["Temperature_Mean"], geo_df_sorted["R0_confidence_interval_Lower"], color="g")
```

```python
geo_df_sorted = geo_df.astype({"Absolute_Humidity":"double","R0_mean":"double","R0_confidence_interval_Upper":"double","R0_confidence_interval_Lower":"double"}).sort_values(by=["Absolute_Humidity"])
plt.title("R0 relation")
plt.ylabel("R0")
plt.xlabel("Absolute_Humidity")
plt.style.use('ggplot')
plt.scatter(geo_df_sorted["Absolute_Humidity"], geo_df_sorted["R0_mean"], color="y")
plt.scatter(geo_df_sorted["Absolute_Humidity"], geo_df_sorted["R0_confidence_interval_Upper"], color="r")
plt.scatter(geo_df_sorted["Absolute_Humidity"], geo_df_sorted["R0_confidence_interval_Lower"], color="g")
```

```python
geo_df_sorted = geo_df.astype({"Lat":"str","R0_mean":"double"}).sort_values(by=["Lat"])
plt.title("R0 relation")
plt.ylabel("R0")
plt.xlabel("Lat")
plt.style.use('ggplot')
plt.scatter(geo_df_sorted["Lat"], geo_df_sorted["R0_mean"], color="y")
plt.scatter(geo_df_sorted["Lat"], geo_df_sorted["R0_confidence_interval_Upper"], color="r")
plt.scatter(geo_df_sorted["Lat"], geo_df_sorted["R0_confidence_interval_Lower"], color="g")
```

```python
geo_df_sorted = geo_df.astype({"Long":"str","R0_mean":"double"}).sort_values(by=["Long"])
plt.title("R0 relation")
plt.ylabel("R0")
plt.xlabel("Long")
plt.style.use('ggplot')
plt.scatter(geo_df_sorted["Long"], geo_df_sorted["R0_mean"], color="y")
plt.scatter(geo_df_sorted["Long"], geo_df_sorted["R0_confidence_interval_Upper"], color="r")
plt.scatter(geo_df_sorted["Long"], geo_df_sorted["R0_confidence_interval_Lower"], color="g")
```

```python
geo_df_sorted = geo_df.astype({"PopulationDensity_2014":"str","R0_mean":"double"}).sort_values(by=["PopulationDensity_2014"])[:-1]
geo_df_sorted.fillna(geo_df_sorted.mean())
plt.title("R0 relation")
plt.ylabel("R0")
plt.xlabel("PopulationDensity_2014")
plt.style.use('ggplot')
plt.scatter(geo_df_sorted["PopulationDensity_2014"], geo_df_sorted["R0_mean"], color="y")
plt.scatter(geo_df_sorted["PopulationDensity_2014"], geo_df_sorted["R0_confidence_interval_Upper"], color="r")
plt.scatter(geo_df_sorted["PopulationDensity_2014"], geo_df_sorted["R0_confidence_interval_Lower"], color="g")
```
